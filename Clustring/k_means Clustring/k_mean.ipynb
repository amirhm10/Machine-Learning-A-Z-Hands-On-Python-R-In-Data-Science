# K-Means Clustring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()
```

```python
x = df.iloc[:, [3, 4]].values
x.shape
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    model = KMeans(n_clusters = i, init = "k-means++", max_iter = 300, n_init = 10)
    model.fit(x)
    wcss.append(model.inertia_)
plt.plot(range(1, 11), wcss)
plt.scatter(range(1, 11), wcss, color="red", marker = "+")
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 5, init = "k-means++", max_iter = 300, n_init = 10)
y_kmeans = model.fit_predict(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], c = y_kmeans, cmap="coolwarm")
plt.scatter(model.cluster_centers_[:, 0],model.cluster_centers_[:, 1], color = "yellow", s=300)
plt.show()
```
